## Dictionary

```python
import pandas as pd
sale = [{'account':'Jones LLC','Jan':150,'Feb':200,'Mar':140},
        {'account':'Alphs Co','Jan':200, 'Feb':200,'Mar':215},
        {'account':'Blue Inc', 'Jan':50,'Feb':210,'Mar':95}]
Dictionary = pd.DataFrame(sale)
```

```python
print(Dictionary)
```

## List

```python
sale = [('Jones LLC',150,200,140),
        ('Alphs Co',200, 200,215),
        ('Blue Inc',210,90,95)]
c1 = ['account','Jan','Feb','Mar']
L2 = pd.DataFrame(sale,columns=c1)
```

```python
print(L2)
```

## contumn oriented from list

```python
sale = {'account':['Jones LLC','Alpha co','Blue Inc'],
                'Jan':[150,200,50],
                 'Feb':[200,210,90],
                  'Mar':[140,215,95]}
contumm = pd.DataFrame(sale)
```

```python
print(contumm)
```

## contumm oriented from item

```python
sale = [('account',['Jones LLC', 'Alpha co', 'Blue Inc']),
          ('Jan', [150,200,50]),
          ('Feb', [200,210,90]),
         ('Mar', [140,215,95]) ]
c2 = pd.DataFrame.from_items (sale)
```

```python
print(c2)
```
