##### สร้าง function ชื่อ smartDict ซึ่งทำหน้าที่:
    -function รับ input เป็น list 2 ตัวที่มีขนาดเท่ากันโดย listแรกเป็น list ของตัวเลข list ตัวที่สองเป็น list ของ text
    -ถ้าค่าใน list แรกเป็นเลขคู่ ให้ค่าใน dict ที่เป็น output มี key เป้นค่าจากlist ตัวที่ 1 และ value จาก listตัวที่ 2
    -ถ้าค่าใน list แรกเป้นเลขคี่ ให้ค่าใน dict ที่เป็น output มี key เป้นค่าจากlist ตัวที่ 2 และ value จาก listตัวที่ 1
    -output ของ function คือ list ที่สร้างขึ้น

```python
def smartDict(k1,k2):
    f = {}
    for i in range(len(k1)):
        if k1 :
            f[k1[i/2 == 0]] = k2[i]
        else:
            f[k2[i]] = k1[i]
    return(f)
```

```python
smartDict([1,2],['apple','banana'])
```
