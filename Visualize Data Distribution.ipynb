## Scatter plot

## numpy สำหรับจัดการข้อมูลที่เป็นตัวเลข

```python
import pandas
import numpy as np
```

```python
irisdata = pandas.read_csv('C:\\Users\\User\\Desktop\\iris.data',header = None)
irisdata
```

## iloc[A:50,:-1] หมายถึงตัดตั้งแต่ 0 ถึง 50

```python
setosa_np = irisdata.iloc[:50,:-1].to_numpy()
setosa_np[:5]
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1])
```

## เพิ่มขนาดให้จุดโดยกำหนด s

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=60)
```

#### s=30*setosa_np[:,2] เพิ่มขนาดเฉพาะบางจุด โดยกำหนดให้ 30*setosa_np[:,2] ช่อง '30*setosa_np[:,2]'มีขนาด 30 อันเดียว

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2])
```

## ช่องที่ 4 แทนสี โดยกำหนด c='k'    C M Y K

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=40*setosa_np[:,2],c='C')
```

## กำหนดเองว่าจะใช้สีอะไรที่ค่าเท่าไหร่

```python
print(f'Max = {max(setosa_np[:,3])},min = {min(setosa_np[:,3])}')
```

```python
colors = []
for i in setosa_np[:,3]:
    if i < 0.5:
        colors.append('b')
    elif i < 0.3:
        colors.append('g')
    elif i < 0.45:
        colors.append('y')
    else:
        colors.append('r')
colors   
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=40*setosa_np[:,2],c= colors )
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=40*setosa_np[:,2],c= setosa_np[:,2] )
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=40*setosa_np[:,2],c= setosa_np[:,2],cmap='YlOrBr')
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=40*setosa_np[:,2],c= np.array([[0.,0.9,0.1]]))
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=40*setosa_np[:,2],c= setosa_np[:,2],cmap='YlOrBr',marker = 'p')
```
